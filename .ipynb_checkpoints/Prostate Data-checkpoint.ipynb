```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
```

Loading data and splitting it into response vector y and matrix of attributes X.

```python
with open(r"prostate.data.txt") as f:
    headers = f.readline().rstrip('\n').split()
    
X = np.loadtxt(r"prostate.data.txt", skiprows = 1)
Y = X[:, -1]
X = X[:, 0:-1]
```

Choose the first 50 patients as the tranining data. And the remaining will be used as testing data

```python
Ytrain, Ytest = Y[0:50], Y[50:]
Xtrain, Xtest = X[0:50], X[50:]
```

Set both variables to have zero mean and standardize the input variables to have unit variance.

```python
Xbar = np.mean(Xtrain, axis = 0)
Xstd = np.std(Xtrain, axis = 0)

Ybar = np.mean(Ytrain)

Ytrain = Ytrain - Ybar
Xtrain = (Xtrain - Xbar) / Xstd
```

As we are dealing with multicollinearity(collinearity) where each predictor variable in a multiple regression model can be linearly predicted from the others with a substantial degree of accuracy, in this situation the coefficient estimates of the multiple regression may change erratically in response to small changes in the model or the data. 

Thus, we will now construct a model using ridge regression as remedial measure taken to alleviate multicollinearity and to predict the 9th variable as a linear combination of the other 8

Computed the ridge regression solutions for a range of regularizers(δ2). Plotted the values of each θ in the
y-axis against δ2 in the x-axis. This set of plotted values is known as a regularization path.

```python
def ridge_auto(Xtrain, Ytrain, d2):   
    model = Ridge(alpha = d2)
    model.fit(Xtrain, Ytrain)
    return model.coef_

def ridge_manual(Xtrain, Ytrain, d2):
    return np.dot(np.dot(np.linalg.inv(np.dot(Xtrain.T, Xtrain) + d2 * np.identity(8)), Xtrain.T), Ytrain)


# regularizerRange = np.linspace(start = 10 ** (-2), stop = 10 ** 3, num = 300)
regularizerRange = np.array([ 10 ** x for index, x in np.ndenumerate(np.linspace(-3.0, 6.0, 10)) ])
regularizerIndices = np.arange(0, regularizerRange.shape[0], 1)
parameters = np.zeros(shape=(regularizerRange.shape[0], Xtrain.shape[1]))

for index, value in np.ndenumerate(regularizerRange):
    parameters[index] = ridge_manual(Xtrain, Ytrain, value)
    
print(ridge_auto(Xtrain, Ytrain, value))
print((Xtrain.T*(1.0 ** (-2))*(Ytrain - Xtrain*)))
                                     
plt.title("Regularization path for ridge regression", pad = 30)
plt.xlabel("δ^2 - regularization parameter")
plt.ylabel("Θ")
    
plt.plot(regularizerIndices, parameters)
plt.xticks(regularizerIndices, regularizerRange)
plt.show()
```

For each computed value of θ, compute the train and test error. Standardizing the test data with the same means and standard deviations as above (Xbar, Xstd, ybar) before you can make a prediction and compute your test error.

```python
len = parameters.shape[0]

def error_prediction(Xcomp, Ycomp):
    Ypred = np.zeros(shape=(len, Xcomp.shape[0]))
    for index in range(len):
        Ypred[index] = Ybar + np.dot(Xcomp, parameters[index])
    
    predError = np.zeros(shape=(len))
    for index in range(len):
        predError[index] = np.linalg.norm(Ycomp - Ypred[index]) / np.linalg.norm(Ycomp)
    
    return predError

plt.title("Relative error of the ridge estimator against regularization parameter δ^2", pad = 30)
plt.xlabel("δ^2 - regularization parameter")
plt.ylabel("||Y - XΘ||2  / ||Y||2")

# Standardize the test/train data with the same means and standard deviations used before
plt.plot(regularizerIndices, error_prediction((Xtest - Xbar) / Xstd, Ytest))
plt.plot(regularizerIndices, error_prediction(Xtrain, Ytrain + Ybar))
plt.xticks(regularizerIndices, regularizerRange)
plt.show()

print(parameters[4])
```

Choosing a value of 4.0-6.0 δ^2 as the regularization parameter using cross-validation as the train/test data comes from the same source of data, because we want to do well in both tests and choose a not so big and not so small regularization parameter to avoid overfitting.
